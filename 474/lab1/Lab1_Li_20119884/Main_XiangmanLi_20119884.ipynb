```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import sklearn
print(cv2.__version__)
print(np.__version__)
print(sklearn.__version__)
```

```python
print("Hello World!")
img = cv2.imread("lena.png")
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Lena Color")
```

```python
plt.imshow(gray, cmap = 'gray')
plt.title("Lena Grayscale")
```

```python
print(img.shape)
print(img.size)
print(img.dtype)
```

```python
himg = np.zeros([512,512,3],np.uint8)
x=y=d=0
for x in range (0,512):
    for y in range (0,512):
        for d in range(0,3):
            himg[x,y,d] = img[x,512-y-1,d]

plt.imshow(cv2.cvtColor(himg,cv2.COLOR_BGR2RGB))
plt.title("Lena Color horizontal flip")
```

```python
vimg = np.zeros([512,512,3],np.uint8)
x=y=d=0
for x in range (0,512):
    for y in range (0,512):
        for d in range(0,3):
            vimg[x,y,d] = img[512-x-1,y,d]

plt.imshow(cv2.cvtColor(vimg,cv2.COLOR_BGR2RGB))
plt.title("Lena Color vertical flip")
```

```python
nimg = np.zeros([512,512,3],np.uint8)
x=y=d=0
for x in range (0,512):
    for y in range (0,512):
        for d in range(0,3):
            (b,g,r)=img[x,y]
            b=255-b
            g=255-g
            r=255-r
            nimg[x,y]=(b,g,r)
                
plt.imshow(cv2.cvtColor(nimg,cv2.COLOR_BGR2RGB))
plt.title("Lena Color negated")
```

```python
nimg = np.zeros([512,512,3],np.uint8)
x=y=d=0
for x in range (0,512):
    for y in range (0,512):
        for d in range(0,3):
            (b,g,r)=img[x,y]
            b=255-b
            g=255-g
            r=255-r
            nimg[x,y]=(b,g,r)
                
plt.imshow(cv2.cvtColor(nimg,cv2.COLOR_BGR2GRAY), cmap = 'gray')
plt.title("Lena Gray negated")
```

```python
bimg = np.zeros([512,512,3],np.uint8)
x=y=d=0
for x in range (0,512):
    for y in range (0,512):
        for d in range(0,3):
            (b,g,r)=img[x,y]
            bimg[x,y]=(b,0,0)
                
plt.imshow(cv2.cvtColor(bimg,cv2.COLOR_BGR2RGB))
plt.title("Lena blue image")
```

```python
gimg = np.zeros([512,512,3],np.uint8)
x=y=d=0
for x in range (0,512):
    for y in range (0,512):
        for d in range(0,3):
            (b,g,r)=img[x,y]
            gimg[x,y]=(0,g,0)
                
plt.imshow(cv2.cvtColor(gimg,cv2.COLOR_BGR2RGB))
plt.title("Lena Green image")
```

```python
rimg = np.zeros([512,512,3],np.uint8)
x=y=d=0
for x in range (0,512):
    for y in range (0,512):
        for d in range(0,3):
            (b,g,r)=img[x,y]
            rimg[x,y]=(0,0,r)
                
plt.imshow(cv2.cvtColor(rimg,cv2.COLOR_BGR2RGB))
plt.title("Lena Red image")
```

```python
mimg = np.zeros([512,512,3],np.uint8) 
x=y=0
for x in range (0,512):
    for y in range (0,512):
            (b1,g1,r1)=bimg[x,y]
            (b2,g2,r2)=gimg[x,y]
            (b3,g3,r3)=rimg[x,y]
            mb=b1+b2+b3
            mg=g1+g2+g3
            mr=r1+r2+r3
            mimg[x,y]=(mb,mg,mr)

plt.imshow(cv2.cvtColor(mimg,cv2.COLOR_BGR2RGB)) 
plt.title("Merged lena image")   
```
